**Step 2.** Use the Twitter Streaming API to download tweets containing the keyword “michigan”. Set the time limit to 30 seconds and store the results into a file named michigan.json. 

```python
import time
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream

C_KEY = 'B9TKXEWyZb7EgxWmfq2Vgke0p'
C_SECRET = 'mUbj1rrUKlYjMGE655bzygP0eY3lS5ihjRjIp3BozTOkkt5yJ2'
A_TOKEN_KEY = '1216914346741325825-A03s8GoafxclYUZutGhroosJ2dhpUH'
A_TOKEN_SECRET = 'Muku809iLWzEite4q3iM0YWZ1QQ2cgHrNPbz0xb7pN5V2'

# Create a StreamListener class 
class MyListener(StreamListener):

    def __init__(self, time_limit=30):
        self.start_time = time.time()
        self.limit = time_limit
        self.outFile = open('michigan.json', 'w')
        super(MyListener, self).__init__()
        
    def on_data(self, data):
        if (time.time() - self.start_time) < self.limit:
            self.outFile.write(data.strip())
            self.outFile.write('\n')
            return True
        else:
            self.outFile.close()
            return False
        
    def on_error(self, status):
        print(status)
        
auth = OAuthHandler(C_KEY,C_SECRET)
auth.set_access_token(A_TOKEN_KEY,A_TOKEN_SECRET)
myStream = Stream(auth, MyListener(time_limit= 30 ))
myStream.filter(track=[ 'michigan'])
```
